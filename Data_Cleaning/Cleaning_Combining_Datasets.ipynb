```python
# import pandas as pd

# Function to handle splitting and filling categories
def fill_categories(row):
    split_categories = row.split('|')
    if len(split_categories) < 3:
        return ['No Category'] * 6
    else:
        return [split_categories[0], split_categories[1], split_categories[-2] if len(split_categories) > 2 else None,
                split_categories[-3] if len(split_categories) > 3 else "No Category", split_categories[2], split_categories[-1]]

# Apply the function to create new columns
sales_cleaned[['masterCategory', 'subCategory', 'additionalCategory1', 'additionalCategory2',
    'articleType', 'productType']] = sales_cleaned['category'].apply(fill_categories).apply(pd.Series)
```
