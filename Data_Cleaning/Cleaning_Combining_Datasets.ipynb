```python
electronics_ratings=electronics_ratings.drop(columns="timestamp")
```

```python
sales
```

```python
sales["product_id"].drop_duplicates(inplace=True)
```

```python
sales_final=sales.drop_duplicates(subset=['product_id'], keep=False,ignore_index=True)
```

```python
sales_final
```

```python
sales_cleaned=sales.drop(columns=["review_id","img_link","product_link","user_name","user_id"])
```

```python
product_sales_merge=pd.merge(left=sales_cleaned,right=electronics_ratings,on="product_id",how="outer")
```
