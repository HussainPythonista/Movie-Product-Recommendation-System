```python

Fashion_data_2=pd.read_csv("/content/drive/MyDrive/Movie-Product-Recommendation /Datasets/Fashion Dataset/styles.csv",on_bad_lines='skip')
```

```python
Fashion_data_2
```

```python
# Data augmentation
augmented_data = []

# Define the number of times you want to augment the data to reach 2 lakh (200,000)
augmentation_factor = 200000 // len(Fashion_data_2)

for i in range(augmentation_factor):
    augmented_data.append(Fashion_data_2.apply(lambda x: x.sample(frac=1).reset_index(drop=True)))

# Concatenate the original and augmented data
enlarged_dataset = pd.concat([Fashion_data_2] + augmented_data, ignore_index=True)

# Trim the dataset to exactly 2 lakh rows
enlarged_dataset = enlarged_dataset.head(200000)

# Check the size of the new dataset
print(enlarged_dataset.shape)
```

```python
enlarged_dataset
```

```python
# Shuffle the dataset
shuffled_dataset = enlarged_dataset.sample(frac=1).reset_index(drop=True)
```

```python
# Generate 'rating' column with values between 0.5 and 5 (inclusive), in increments of 0.5
enlarged_dataset['rating'] = np.random.choice(np.arange(0.5, 5.5, 0.5), size=len(enlarged_dataset))

# Check the first few rows of the dataset
enlarged_dataset.head()
```
